```python
import pandas as pd
import matplotlib.pyplot as plt

# List of years and corresponding DataFrames
years = [2020, 2021, 2022, 2023, 2024]
dataframes = [_2020, _2021, _2022, _2023, _2024]  # Replace with actual DataFrame names

# Create dictionaries to store yearly ride counts
casual_rides_per_year = {}
member_rides_per_year = {}

# Loop through each year and count rides for casual and member
for year, df in zip(years, dataframes):
    casual_count = df[df["member_casual"] == "casual"]["ride_id"].count()
    member_count = df[df["member_casual"] == "member"]["ride_id"].count()

    casual_rides_per_year[year] = casual_count
    member_rides_per_year[year] = member_count

# Convert dictionaries to DataFrames
rides_df = pd.DataFrame({
    "Year": years,
    "Casual Rides": [casual_rides_per_year[year] for year in years],
    "Member Rides": [member_rides_per_year[year] for year in years]
})

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(rides_df["Year"], rides_df["Casual Rides"], marker="o", linestyle="-", color="b", label="Casual Riders")
plt.plot(rides_df["Year"], rides_df["Member Rides"], marker="s", linestyle="--", color="g", label="Annual Members")

# Add labels, title, and legend
plt.xlabel("Year")
plt.ylabel("Number of Rides")
plt.title("Annual Trend of Casual and Member Riders")
plt.xticks(years)  # Ensure x-axis has correct years
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

years = [2020, 2021, 2022, 2023, 2024]  
dataframes = [_2020, _2021, _2022, _2023, _2024]  

# Define seasons
season_labels = ["Winter", "Spring", "Summer", "Fall"]
avg_duration_data = {season: [] for season in season_labels}  # Dictionary for avg durations

for year, df in zip(years, dataframes):
    # Filter for only casual riders
    casual_df = df[df["member_casual"] == "casual"]
    
    # Calculate average ride duration per season
    avg_duration_by_season = casual_df.groupby("season")["duration_minutes"].mean()  # Assuming 'duration_minutes' column exists
    
    # Store durations, ensuring all seasons are represented
    for season in season_labels:
        avg_duration_data[season].append(avg_duration_by_season.get(season, 0))  # Default to 0 if missing

# Plot average ride duration per season
fig, ax = plt.subplots(figsize=(12, 6))

# Different markers for each season
markers = ['o', 's', '^', 'd']
colors = ['b', 'g', 'r', 'purple']  # Different colors for better visibility

# Plot each season separately with value labels
for i, season in enumerate(season_labels):
    ax.plot(years, avg_duration_data[season], marker=markers[i], linestyle='-', color=colors[i], label=season)
    
    # Annotate each point with the corresponding value
    for x, y in zip(years, avg_duration_data[season]):
        ax.text(x, y, f"{y:.1f}", fontsize=10, ha='center', va='bottom', color=colors[i])

# Formatting the plot
ax.set_ylabel("Average Ride Duration (minutes)")
ax.set_xlabel("Year")
ax.set_title("Seasonal Average Ride Duration Per Year (Casual Riders)")
ax.legend(title="Season")
ax.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

highest_ridden_stations = {}

for year, df in zip(years, dataframes):
    # Filter for casual riders in Winter
    winter_casual_df = df[(df["season"] == "Winter") & (df["member_casual"] == "casual")]
    
    # Count rides per start station
    station_counts = winter_casual_df.groupby("start_station_name")["ride_id"].count()
    
    # Find the top 5 most used stations
    highest_stations = station_counts.nlargest(5)
    
    # Store in dictionary
    highest_ridden_stations[year] = highest_stations

# Prepare data for plotting
all_stations = set()
for stations in highest_ridden_stations.values():
    all_stations.update(stations.index)  # Collect unique station names

all_stations = sorted(all_stations)  # Sort for consistent order
x_indexes = np.arange(len(all_stations))  # X-axis positions
bar_width = 0.15  # Width of each bar

# Plot
fig, ax = plt.subplots(figsize=(12, 6))

for i, (year, stations) in enumerate(highest_ridden_stations.items()):
    # Get ride counts for all stations (fill missing stations with 0)
    ride_counts = [stations.get(station, 0) for station in all_stations]
    
    # Plot bars
    ax.bar(x_indexes + i * bar_width, ride_counts, width=bar_width, label=str(year))

# Formatting the plot
ax.set_xticks(x_indexes + bar_width * (len(highest_ridden_stations) / 2))
ax.set_xticklabels(all_stations, rotation=45, ha="right")
ax.set_ylabel("Ride Count")
ax.set_xlabel("Station Name")
ax.set_title("Top 5 Most Ridden Stations in Winter (Casual Riders)")
ax.legend(title="Year")

plt.tight_layout()
plt.show()
```
